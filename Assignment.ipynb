```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig

class BoxParticle:
    def __init__(self, L, N):
        self.L = L
        self.N = N
        self.dx = L / N
        self.x = np.linspace(0, L, N)
        self.hbar = 1.055e-34  # Planck's constant divided by 2π
        self.m = 9.11e-31  # mass of electron
        self.V = np.zeros(N)  # default potential energy function is zero
        
    def potential(self, V):
        if len(V) != self.N:
            raise ValueError("Potential array must have length N")
        self.V = V
        
    def hamiltonian(self):
        H = np.zeros((self.N, self.N))
        for i in range(1, self.N-1):
            H[i, i] = -2 / (self.dx ** 2) + self.V[i]
            H[i, i-1] = 1 / (self.dx ** 2)
            H[i, i+1] = 1 / (self.dx ** 2)
        H[0, 0] = -2 / (self.dx ** 2) + self.V[0]
        H[0, 1] = 1 / (self.dx ** 2)
        H[-1, -1] = -2 / (self.dx ** 2) + self.V[-1]
        H[-1, -2] = 1 / (self.dx ** 2)
        return H
    
    def solve(self):
        H = self.hamiltonian()
        E, psi = eig(H)
        idx = E.argsort()
        self.E = E[idx]
        self.psi = psi[:, idx]
        
    def plot_wave_functions(self, num_functions=5):
        fig, ax = plt.subplots()
        for i in range(num_functions):
            ax.plot(self.x, self.psi[:, i], label=f"n={i+1}")
        ax.set_xlabel("x")
        ax.set_ylabel("ψ(x)")
        ax.legend()
        plt.show()
        
    def plot_energy_levels(self):
        fig, ax = plt.subplots()
        ax.plot(range(1, len(self.E)+1), self.E, 'bo')
        ax.set_xlabel("n")
        ax.set_ylabel("Energy (J)")
        ax.set_title("Particle in a one-dimensional box")
        plt.show()
```

```python
pib = BoxParticle(L=1e-9, N=1000)
pib.potential(np.zeros(pib.N))
pib.solve()
pib.plot_wave_functions(num_functions=3)
pib.plot_energy_levels()
```
